```python
import numpy as np
from numpy import diff, log, exp, cumsum
import matplotlib.pyplot as plt
import random

def randomWalk(steps, prob):
    '''
    Computes a random walk with a given number of steps, and a probability
    of up movement.
    Function plots the random walk and prints minimum and maximum values
    '''
    position = 0
    walk = [position]
    
    for i in range(steps):
        if np.random.binomial(1, prob) == 1:
            step = 1
        else:
            step = -1
        position += step
        walk.append(position)
    
    plt.plot(walk)
    #Upgrade the graph
    print('The maximum value is', max(walk))
    print('The minimum value is', min(walk))

randomWalk(10000, 0.5)
```

```python
def nRandomWalks(n, steps, prob):
    '''
    '''
    
    numbers = np.random.randint(0, 2, size=(steps, n))
    movements = np.where(numbers == 0, 1, -1)
    values = np.cumsum(movements, axis=0)
    plt.xlabel('Time')
    plt.ylabel('Value')
    ax = plt.plot(values)
    
nRandomWalks(4, 1000, 0.5)
```

```python
def multiRandomWalk(steps, initial_value):
    '''
    '''
    
    random_numbers = np.random.normal(size=steps) * 0.005
    multipliers = 1 + random_numbers
    values = initial_value * np.cumprod(multipliers)
    plt.xlabel('t')
    plt.ylabel('y')
    ax = plt.plot(values)
    
multiRandomWalk(10000, 100)
```

```python
def multLogRandomWalk(initial_value, time, volatility):
    values = np.random.normal(size=time)*volatility
    y = initial_value * exp(cumsum(values))
    plt.xlabel('Time')
    plt.ylabel('Y')
    ax = plt.plot(y)

multLogRandomWalk(100, 1000, 0.001)
```
